```python
import pandas as pd
%matplotlib inline
df = pd.read_csv('q1_q17_dr25_koifpp_2020.05.02_01.07.33.csv',skiprows=26)
koi = pd.read_csv('cumulative_2020.05.02_01.18.44.csv',skiprows=66)
df
koi
```

```python
beb = df[df.fpp_prob_beb>0.5]
beb.fpp_steff.hist(bins=50)
```

```python
beb_teff = beb[beb.fpp_steff<6500]
beb_teff
```

```python
beb = pd.merge(beb_teff,koi,on='kepid')
beb
```

```python
beb_kep = beb[(beb.koi_kepmag>12)&(beb.koi_kepmag<16)]
```

```python
import astropy.units as u
from astropy.coordinates import SkyCoord
import matplotlib.pyplot as plt
radec = SkyCoord(beb_kep.ra*u.degree,beb_kep.dec*u.degree)
beb_kep['l']=radec.transform_to('galactic').l
beb_kep['b']=radec.transform_to('galactic').b
```

```python
beb_kep.dropna(subset=['koi_zmag'],inplace=True)
beb_kep
```

```python
beb_kep.koi_steff.hist(bins=25)
```

```python
beb_kep.koi_kepmag.hist(bins=50)
```

```python
beb_kep.to_csv('beb_sample.csv',index=False)
```

```python
beb_cand = beb_kep[beb_kep.koi_pdisposition=='CANDIDATE']
beb_cand
```

```python
beb_cand.columns
```

```python
tmp = beb_cand.copy()
data = pd.DataFrame()
data['host_name'] = tmp['kepler_name']
data['koi'] = tmp['kepoi_name']
data['RA'] = tmp['ra']
data['DEC'] = tmp['dec']
data['Teff'] = tmp['koi_steff']
data['Teff_err'] = tmp['koi_steff_err1']
data['g_mag'] = tmp['koi_gmag']
data['r_mag'] = tmp['koi_rmag']
data['i_mag'] = tmp['koi_imag']
data['z_mag'] = tmp['koi_zmag']
data['J_mag'] = tmp['koi_jmag']
data['H_mag'] = tmp['koi_hmag']
data['K_mag'] = tmp['koi_kmag']
data['Kepler_mag'] = tmp['koi_kepmag']
data['period'] = tmp['koi_period']
data['transit_depth'] = tmp['koi_depth']
data['l'] = tmp['l']
data['b'] = tmp['b']
data.to_csv('beb_sample.csv',index =False)
```
