```python
plt.figure(figsize=[20,7])

plt.subplot(121)
bp = 'plpop_H_Giants'
qua = pl_df[bp+'_qua'].values
med = pl_df[bp+'_med'].values
tqua = pl_df[bp+'_tqua'].values
y_err = np.array((med-qua,tqua-med))
plt.errorbar(x=pl_df.plpop_teff_0,y=med,yerr=y_err,fmt='.')
plt.subplot(122)
band = ['g','r','i','z','J','H','K']
for b in band:
    bp = 'plpop_'+b+'_Giants'
    qua = pl_df[bp+'_qua'].values
    med = pl_df[bp+'_med'].values
    tqua = pl_df[bp+'_tqua'].values
    y_err = np.array((med-qua,tqua-med))
    plt.errorbar(x=pl_df.Teff,y=med,yerr=y_err,fmt='.',label=b)    
plt.legend()
#plt.title('Companion Transiting Planet: Giants multicolor difference',fontsize=15)
plt.yscale('log')

plt.xlabel('Primary Effective Tempreature')

plt.savefig('teff.pdf',bbox_inches='tight')
```

```python
band = ['g','r','i','z','J','H','K']
planet_class = ['Earth','Super_Earth','Small_Neptunes','Large_Neptunes','Jupiters','Giants']
bp = 'plpop_H_Giants'
qua = pl_df[bp+'_qua'].values
med = pl_df[bp+'_med'].values
tqua = pl_df[bp+'_tqua'].values
y_err = np.array((med-qua,tqua-med))
plt.figure(figsize=(10,6))
plt.errorbar(x=pl_df.plpop_teff_0,y=med,yerr=y_err,fmt='.')
plt.title('Companion Transiting Planet: Giants H-Kepler difference',fontsize=15)
#plt.yscale('log')
```
